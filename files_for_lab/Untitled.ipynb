```python
#1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2
file_path = 'we_fn_use_c_marketing_customer_value_analysis.csv'

customer_df = pd.read_csv(file_path)
customer_df
```

```python
#3
customer_df.head()
customer_df.shape 
customer_df.info()  
```

```python
#4
customer_df = customer_df.rename(columns=lambda x: x.lower().replace(' ', '_'))
print(customer_df.columns)
```

```python
#5
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'], format='%Y-%m-%d')
```

```python
#6
numerical_df = customer_df.select_dtypes(include='number')
display(numerical_df)
```

```python
#7
def differentiate_variables(df):
    continuous_vars = []
    discrete_vars = []

    for col in df.columns:
        if df[col].nunique() > 15:
            continuous_vars.append(col)
        else:
            discrete_vars.append(col)

    return continuous_vars, discrete_vars

continuous_vars, discrete_vars = differentiate_variables(numerical_df)


continuous_df = numerical_df[continuous_vars]
discrete_df = numerical_df[discrete_vars]
```

```python
print("Continuous Variables:")
display(continuous_df.head())
```

```python
print("\nDiscrete Variables:")
display(discrete_df.head())
```

```python
#8
import seaborn as sns

correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
#9
def plot_discrete_variables(df):
    for col in df.columns:
        plt.figure(figsize=(8, 6))
        sns.countplot(x=col, data=df)
        plt.title(f'Distribution of {col}')
        plt.show()
```

```python
def plot_continuous_variables(df):
    for col in df.columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.show()
```

```python
plot_discrete_variables(discrete_df)
```

```python
plot_continuous_variables(continuous_df)
```
