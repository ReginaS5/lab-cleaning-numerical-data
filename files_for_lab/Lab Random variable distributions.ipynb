```python
df = pd.read_csv('numerical.csv')
#1
df
```

```python
#2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, expon


column_name = 'customer_lifetime_value'

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df[column_name], kde=True, ax=ax1, label='Data Histogram')
sns.lineplot(x=np.linspace(min(df[column_name]), max(df[column_name]), 100),
             y=norm.pdf(np.linspace(min(df[column_name]), max(df[column_name]), 100),
                        np.mean(df[column_name]), np.std(df[column_name])),
             color='red', label='Fitted Normal Distribution', ax=ax1)
ax1.set_title('Normal Distribution Fit')

sns.histplot(df[column_name], kde=True, ax=ax2, label='Data Histogram')
sns.lineplot(x=np.linspace(min(df[column_name]), max(df[column_name]), 100),
             y=expon.pdf(np.linspace(min(df[column_name]), max(df[column_name]), 100),
                         np.mean(df[column_name]), np.std(df[column_name])),
             color='red', label='Fitted Exponential Distribution', ax=ax2)
ax2.set_title('Exponential Distribution Fit')

plt.tight_layout()
plt.show()
```

```python
#3

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm, expon

column_name = 'customer_lifetime_value'

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df[column_name], kde=True, ax=ax1, label='Original Data Histogram')
sns.lineplot(x=np.linspace(min(df[column_name]), max(df[column_name]), 100),
             y=norm.pdf(np.linspace(min(df[column_name]), max(df[column_name]), 100),
                        np.mean(df[column_name]), np.std(df[column_name])),
             color='red', label='Fitted Normal Distribution', ax=ax1)
ax1.set_title('Original Distribution')

log_transformed_data = np.log(df[column_name] + 1)  # Adding 1 to avoid issues with log(0)
sns.histplot(log_transformed_data, kde=True, ax=ax2, label='Log-Transformed Data Histogram')
sns.lineplot(x=np.linspace(min(log_transformed_data), max(log_transformed_data), 100),
             y=norm.pdf(np.linspace(min(log_transformed_data), max(log_transformed_data), 100),
                        np.mean(log_transformed_data), np.std(log_transformed_data)),
             color='red', label='Fitted Normal Distribution', ax=ax2)
ax2.set_title('Log-Transformed Distribution')


plt.tight_layout()
plt.show()
```
